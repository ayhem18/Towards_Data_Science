## Part 1. Solving the unconstrained optimization problem

Consider the problem of empirical risk minimization:
\begin{equation}
\min_{w \in \mathbb{R}^d} \frac{1}{n} \sum\limits_{i=1}^n \ell (g(w, x_i), y_i) + \frac{\lambda}{2} \| w \|^2_2,
\end{equation}
where $\ell$ is the loss function, $g$ is the model, $w$ is the model parameters, $\{x_i, y_i\}_{i=1}^n$ is the data sample from feature vectors $x_i$ and labels $y_i$, $\lambda > 0$ is the regularization parameter.

We use the linear model $g(w, x) = w^T x$ and the logistic/sigmoidal loss function: $\ell(z,y) = \ln (1 + \exp(-yz))$ (__Important: $y$ must take values $-1$ or $1$__). The resulting problem is called a logistic regression problem.

### __а). (15 points)__ Write out the gradient and the Hessian for this problem. 

Prove that the problem is $\mu$-strongly convex and has $L$-Lipschitz gradient with $\mu = \lambda$ and $L = \lambda + \frac{1}{4n} \sum_{i=1}^n \| x_i\|^2_2$. 

Let $L_s$ be the problem's function. Let's define $L_s = L_1 + L_2$, where
\begin{align}
L_1 &= \frac{1}{n}\sum_{i=1}^{n} l(<w, x_i>, y_i) \\
L_2 &= \frac{\lambda}{2} \| w \|_2 ^ 2
\end{align}
$L_1$ is a sum, so let's define the derivative of each term.

Let's denote $y_i$ by $y$, $x_i$ by $x$,and $z = <y.x, w>$

Then, we have:
\begin{align}
dz = <yx, dw>
\end{align}
Defining $h = \ln(1 + \exp(-t))$, then $h^{'} = \frac{-1}{1 + e^t}$ and:

\begin{align}
f &= h(z)  && \text{$f$ is a single term in the sum}\\
df &= h'(z) \cdot dz \\
df &=  \frac{-1}{1 + e^z} \cdot <y.x, dw> \\
df &=  \frac{- <y.x, dw>}{1 + e^{<y.x, w>}}  \\
\implies \nabla f &= -\frac{y \cdot x}{1 + e^{<y \cdot x, w>}} \\
\implies \nabla L_1 &= \sum_{i = 1}^n -\frac{y_i \cdot x_i}{1 + e^{<y_i \cdot x_i, w>}}
\end{align}

As for $L_2$, we have:
\begin{align}
d(L_2) &= \lambda \cdot <w, dw> \\
\implies \nabla L_s &= - \frac{1}{n}\sum_{i = 1}^n \frac{y_i \cdot x_i}{1 + e^{<y_i \cdot x_i, w>}} + \lambda \cdot w
\end{align}


* 2nd order derivative: Hessian matrix:

I will use the same approach as in the first part.
\begin{align}
d^2(f) &= <d (\nabla f), dw_1> \\
d (\nabla f) &= - \frac{(1 + e^z) \cdot d(y \cdot x) - d(1 + e^z) \cdot y \cdot x}{(1 + e^z)^2} \\
d(\nabla f) &= \frac{e ^ z \cdot <y \cdot x, dw> \cdot y \cdot x }{(1 + e^z)^2}&& \text{differentiation with respect to $w \implies d(y \cdot x) = 0$} \\
\implies d(\nabla f) &= \frac{e ^ z \cdot x\cdot <x, dw>}{(1 + e^z)^2} && \text{using $y \in \{1, -1\}$} \\
\implies d(\nabla f) &= \frac{e ^ z \cdot x\cdot x^T \cdot dw}{(1 + e^z)^2} \\
\implies d^2(f) &= <\frac{e ^ z \cdot x\cdot x^T \cdot dw}{(1 + e^z)^2}, dw_1> \\
\implies d^2(f) &= <\frac{e ^ z \cdot x\cdot x^T}{(1 + e^z)^2} \cdot dw_1, dw> && \text{using $<Ab, c> = <A^Tc, b>$} \\
\implies \nabla ^ 2 f &= \frac{e ^ {<y \cdot x, w>} \cdot x\cdot x^T}{(1 + e^{<y \cdot x, w>})^2} \\
\implies \nabla ^ 2 L_1 &= \frac{1}{n} \sum_{i=1}^n \frac{e ^ {<y_i \cdot x_i, w>} \cdot x_i\cdot x_i^T}{(1 + e^{<y_i \cdot x_i, w>})^2}
\end{align}

As for $\nabla ^ 2 L_2 = λ$. Therefore,
$$ \nabla ^ 2 L_s = \frac{1}{n} \sum_{i=1}^n \frac{e ^ {<y_i \cdot x_i, w>} \cdot x_i\cdot x_i^T}{(1 + e^{<y_i \cdot x_i, w>})^2} + λ \cdot I_d  
$$


* Let's prove strong convexity. 

It is known that a differentiable function (such as our function), is strongly convex, only and only if: 

$$
\nabla^2 f(x) \geq \mu I.
$$

We can see that for any vector $x$, the matrix $x \cdot x^T$ is positive definte, since:

\begin{align}
 a^T \cdot (x \cdot x^T) \cdot a &= b^T \cdot b \geq 0&& \text{where $b = x ^ T \cdot a$}
\end{align}

and since the exponential function is always positive (for any argument), the following matrix: 

$$
\frac{e ^ {<y_i \cdot x_i, w>} \cdot x_i\cdot x_i^T}{(1 + e^{<y_i \cdot x_i, w>})^2}
$$

is also positive definite. Using the fact that the sum of positive definite matrices, is also a positive definite matrix. 

We have the following: 

$$
\frac{1}{n} \sum_{i=1}^n \frac{e ^ {<y_i \cdot x_i, w>} \cdot x_i\cdot x_i^T}{(1 + e^{<y_i \cdot x_i, w>})^2} >= 0 \\ 

\implies \nabla ^ 2 L_s -  λ \cdot I_d >= 0
$$

This proves that the problem is $\lambda$ strongly convex.



* Let's prove that $L = \lambda + \frac{1}{4n} \sum_{i=1}^n \| x_i\|^2_2$ 

For a twice-differentiable function $f$, we know a function has $L$-Lipschitz gradient if and only if:

$$ L \cdot I_d  \geq  \nabla^2 f $$

\begin{align}
L \cdot I_d = \lambda \cdot I_d + \frac{1}{4n} \sum_{i=1}^n \| x_i\|^2_2 \cdot I_d &\geq \frac{1}{n} \sum_{i=1}^n \frac{e ^ {<y_i \cdot x_i, w>} \cdot x_i\cdot x_i^T}{(1 + e^{<y_i \cdot x_i, w>})^2} + λ \cdot I_d \\
\iff \frac{1}{4} \sum_{i=1}^n \| x_i\|^2_2 \cdot I_d &\geq  \sum_{i=1}^n \frac{e ^ {<y_i \cdot x_i, w>} \cdot x_i\cdot x_i^T}{(1 + e^{<y_i \cdot x_i, w>})^2} 
\end{align}

* To prove the last inequality we proceed as follows: 

1. It is known that for any $x \in \mathbb{R}$, we have $(1 + x)^2 \geq 4x$ (the difference is a square) which is equivalent to: 

$$ \frac{1}{4} \geq \frac{x}{(1 + x)^2}$$ 

replacing $x$ by $e^{<y_i \cdot x_i, w>}$ we have the following: 

$$ \frac{1}{4} \geq \frac{e^{<y_i \cdot x_i, w>}}{(1 + e^{<y_i \cdot x_i, w>})^2} $$

Let's prove the following inequality, for any vector $x$: 

$$
\|x\|^2  \cdot I_d \geq x \cdot x ^ T
$$

\begin{align}
y^T \cdot (\|x\|^2  \cdot I_d - x \cdot x ^ T) \cdot y &= \|x\|^2 \cdot y ^ T\cdot y  - y^T x \cdot x ^ T \cdot y  && \text{where $y, x$ are any vectors}\\
y^T \cdot (\|x\|^2  \cdot I_d - x \cdot x ^ T) \cdot y &= \|x\|^2 \cdot \|y \| ^ 2 - <x, y> ^ 2 \\
y^T \cdot (\|x\|^2  \cdot I_d - x \cdot x ^ T) \cdot y &\geq 0 && \text{using Cauchy Schwarz inequality}
\end{align}

This is true for any vectors $y, x$. Which means that for any $x$, the matrix: 

$$ \|x\|^2  \cdot I_d - x \cdot x ^ T ~~ \text{is semi positive definite }\\
\implies \|x\|^2  \cdot I_d \geq x \cdot x ^ T
$$

Using this fact and the inequality above, we have:

$$ 
\frac{1}{4} \|x_i\|^2  \cdot I_d \geq \frac{e^{<y_i \cdot x_i, w>}}{(1 + e^{<y_i \cdot x_i, w>})^2} x \cdot x ^ T ~~ \forall x_i
$$
and since the sum of (Semi) positive definite matrices is also a (semi positive definte) matrix: 

$$
\frac{1}{4} \sum_{i=1}^n \| x_i\|^2_2 \cdot I_d \geq  \sum_{i=1}^n \frac{e ^ {<y_i \cdot x_i, w>} \cdot x_i\cdot x_i^T}{(1 + e^{<y_i \cdot x_i, w>})^2} 
$$

proving that for

$$ L  = \lambda + \frac{1}{4n} \sum_{i=1}^n \| x_i\|^2_2 $$  

the problem  has $L$-Lipschitz gradient

```python
import numpy as np
from typing import Tuple, List, Union

def p1_verify_input(X: np.ndarray, y: np.ndarray, w: np.ndarray) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    # first let's make sure the input is as expected: X is expected to have samples as rows
    if X.ndim != 2:
        raise ValueError(f"the training is expected to have at least 2 samples")
    
    num_samples, dim = X.shape 
    # make sure 'y' matches the number of samples
    y = np.expand_dims(y, axis=1) if y.ndim == 1 else y

    if y.shape != (num_samples, 1):
        raise ValueError((f"The number of labels is expected to match the number of samples"
                          f"\nFound: {y.shape} Expected: {(num_samples, 1)}"))

    w = np.squeeze(w)    
    # check 'w' as well:
    w = np.expand_dims(w, axis=-1) if w.ndim == 1 else w

    # make sure the dimensions match
    if w.shape != (dim, 1):
        raise ValueError((f"The weight matrix 'w' is expected as a column vector with length {dim}\n"
                          f"Expected: {(dim, 1)}. Found: {w.shape}"))
    return X, y, w
```

```python
def p1_value_function(X: np.ndarray, 
                   y: np.ndarray,
                   w: np.ndarray, 
                   lam: float) -> float:
    """ This function calculates different values of the function gives it parameters:

    Args:
        X (np.ndarray): The training data
        y (np.ndarray): The labels
        w (np.ndarray): the weights
        lam (float): lambda: the regularization hyper-parameter

    Returns:
        float: The value for the function with parameter 'w'
    """
    # first step is to verify the input
    X, y, w = p1_verify_input(X, y, w)
    
    num_samples, _ = X.shape
    # now as our input has been verified we proceed by defining some useful intermediate variables
    
    A = X * y
    
    return np.mean(np.log(1 + np.exp((- A @ w).astype(np.float64)))) + (lam / 2) * np.linalg.norm(w) ** 2 
```

```python
def p1_gradient_function(X: np.ndarray, 
                      y: np.ndarray, 
                      w: np.ndarray, 
                      lam: float) -> np.ndarray:
    # verify the input
    X, y, w = p1_verify_input(X, y, w)

    A = X * y
    preds = A @ w 

    # preds contains the w^T x_i in each row
    # apply the exponential on each
    # A / (1 + np.exp(preds)) contains y_i * x_i / (1 + e^{w_i^T y_i x_i}) in each row: which a row vector

    # then calculate the mean over axis=0 to get a row vector
    
    dl1 = np.mean(A / (1 + np.exp(preds)), axis=0)

    assert dl1.shape == (X.shape[1], )

    dl = - np.expand_dims(dl1, axis=-1) + lam * w

    assert dl.shape == (X.shape[1], 1), "The gradient is expected to be a column vector."
    
    return dl
```

```python
def p1_hessian_function(X: np.ndarray, 
                   y: np.ndarray, 
                   w: np.ndarray, 
                   lam: float) -> np.ndarray:
    # verify the input
    X, y, w = p1_verify_input(X, y, w)
    num_samples, dim = X.shape

    A = X * y
    preds = (A @ w).astype(np.float64) 
    # preds = preds.astype(np.float64)

    # create the hessian matrix with zero values
    hessian_matrix = np.zeros((dim, dim), dtype=np.float64)

    exp_coeffs = (1 / (1 + np.exp(preds))) - (1 / ((1 + np.exp(preds)) ** 2))

    assert exp_coeffs.shape == (num_samples, 1)

    #iterate through samples
    for i in range(num_samples):
        # first step calculate the x_i * x_i ^ T
        # extract x_i: row vector in code
        x_i = X[i, :]
        
        assert x_i.shape == (dim,) , "The row sample is expected to be 1 dimensional"
        # expand 
        x_i = np.expand_dims(x_i, axis=-1)

        matrix_xi = x_i @ x_i.T
        # add an assert to catch any errors with shapes
        assert matrix_xi.shape == (dim, dim), "Make sure the matrix x_i * x_i ^ T is computed correctly"

        hessian_xi = exp_coeffs[i][0] * matrix_xi
        # time to add the coefficient associated with the matrix_xi
        # hessian_xi = (exp1[i][0] / exp2[i][0]) * matrix_xi
        hessian_matrix += hessian_xi

    # make sure to divide by the number of samples 
    hessian_matrix = hessian_matrix / num_samples + lam * np.eye(dim)
    # make sure the shape of the hessian matrix 
    hessian_matrix.shape == (dim , dim), "Make sure the hessian matrix is of the correct shape"
    return hessian_matrix
```

```python
from numpy.linalg import norm
def x_difference(x1: np.ndarray, x2: np.ndarray) -> float:
    """Computes the euclidean norm of the difference vector 

    Args:
        x1 (np.ndarray): first vector
        x2 (np.ndarray): second vector

    Returns:
        float: returns the norm of the difference
    """
    return norm(x1 - x2, ord=2)

def f_difference(f1: float, f2: float) -> float:
    """returns the absolute difference between 2 values """
    # the expression f_x_k - f_sol is equivalent since the problem is minimization,
    # but the 'abs' function was used to make the method general and not only specific to the given problem
    return abs(f1 - f2)      

# let's define the modes in terms of strings
x_diff = 'x_k+1 - x_k'
f_diff = 'f(x_k+1) - f(x_k)'
normalized_criterion = 'df_xk / df_x0'
x_opt_diff = 'x* - x_k'
```

```python
from tqdm import tqdm
def heavy_ball_method(function: callable, 
                      grad_function: callable, 
                      x_0: np.ndarray, 
                      gamma_k: callable,
                      momentum_k: callable,  
                      K = 10**3, 
                      eps = 10**-5, 
                      mode = normalized_criterion, 
                      return_history: bool = True, 
                      x_sol: np.ndarray=None
                      ):
    
    # the first step is to make sure the 'mode' variable is defined above
    if isinstance(mode , str) and mode not in [f_diff, x_diff, normalized_criterion, x_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))
    
    if mode == x_opt_diff and x_sol is None: 
        raise ValueError(f"using mode = {x_opt_diff} requires passing the solution to the problem")

    # x_p0: stands for the x{k - 1}, x_p1: x{k - 2}
    x_new, x_current, x_previous = x_0, x_0, x_0

    # create a list to save the intermediate points  
    x_history = [x_current]
    # the first criterion is '1' if we are using the normalized criterion and '0' otherwise
    criterion_history = []

    for k in tqdm(range(K)):
        x_previous = x_current
        x_current = x_new
        
        # compute the step size
        gamma = gamma_k(k)
        mom = momentum_k(k)

        # x_{k + 1} = x_k - g_k * nabla(f_xk) + momentum * (x_k - x_{k - 1})
        x_new = x_current - gamma * grad_function(x_current) + mom * (x_current - x_previous) 
        
        if mode == f_diff:
            diff = f_difference(function(x_new), function(x_current))
        
        elif mode == x_diff:
            diff = x_difference(x_new, x_current)
        
        elif mode == normalized_criterion:
            diff = norm(grad_function(x_current), ord=2) / norm(grad_function(x_0), ord=2)
        
        elif mode == x_opt_diff: 
            diff = norm(x_new - x_sol, ord=2)
        else: 
            # the last case is where the criterion is passed as an argument
            diff = mode(x_current)

        # add 'x_current' to the history
        x_history.append(x_current)
        # add the current criterion to history 
        criterion_history.append(diff)
        
        if diff <= eps:
            break
    
    assert len(x_history) == k + 2, f"expected {k + 2} points. Found: {len(x_history)}"

    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

```python
def naesterov_method(function: callable, 
                      grad_function: callable, 
                      x_0: np.ndarray, 
                      gamma_k: callable,
                      momentum_k: callable,  
                      K = 10**3, 
                      eps = 10**-5, 
                      mode = normalized_criterion, 
                      return_history: bool = True, 
                      x_sol: np.ndarray=None):
    
    # the first step is to make sure the 'mode' variable is defined above
    if isinstance(mode , str) and mode not in [f_diff, x_diff, normalized_criterion, x_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))
    
    if mode == x_opt_diff and x_sol is None: 
        raise ValueError(f"using mode = {x_opt_diff} requires passing the solution to the problem")
    
    # set the variables for the parameters 'x', 'y'
    x_current, x_previous, y_current = x_0, x_0, x_0
    # create a list to save the intermediate points  
    x_history = [x_current]
    criterion_history = []

    for k in tqdm(range(K)):
        x_previous = x_current
        
        # compute the step size
        gamma = gamma_k(k)
        mom = momentum_k(k)
        # compute the (k + 1)-th 'x' point
        # compute the (k + 1)-th 'y' point
        x_current = y_current - gamma * grad_function(y_current)
        y_current = x_current + mom * (x_current - x_previous)
        
        if mode == f_diff:
            diff = f_difference(function(x_current), function(x_current))
        
        elif mode == x_diff:
            diff = x_difference(x_current, x_current)
        
        elif mode == normalized_criterion:
            diff = norm(grad_function(x_current), ord=2) / norm(grad_function(x_0), ord=2)
        
        elif mode == x_opt_diff: 
            diff = norm(x_current - x_sol, ord=2)
        else: 
            # the last case is where the criterion is passed as an argument
            diff = mode(x_current)

        criterion_history.append(diff)
        # add 'x_current' to the history
        x_history.append(x_current)
        
        if diff <= eps:
            break

    assert len(x_history) == k + 2, f"expected {k + 2} points. Found: {len(x_history)}"

    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

```python
def gradient_descent(function: callable, 
                     grad_function: callable, 
                     x_0: np.ndarray,                      
                     x_sol: np.ndarray=None,
                     K: int = 10 ** 3,
                     eps: float = 10 ** -5, 
                     mode: str = normalized_criterion,
                     gamma_k: callable = None,                      
                     return_history: bool = False
                     ) -> Union[List[np.ndarray], np.ndarray]:
    # the first step is to make sure the 'mode' variable is defined above
    if isinstance(mode , str) and mode not in [f_diff, x_diff, normalized_criterion, x_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))
    
    if mode == x_opt_diff and x_sol is None: 
        raise ValueError(f"using mode = {x_opt_diff} requires passing the solution to the problem")
    x_current, x_previous = x_0, x_0
    x_history = [x_current]
    criterion_history = []

    for k in tqdm(range(K)):
        # save the k-th point
        x_previous = x_current

        gamma = gamma_k(k)

        # compute the (k+1)-th point
        x_current = x_current - gamma * grad_function(x_current)  

        if mode == f_diff:
            diff = f_difference(function(x_current), function(x_previous))
        
        elif mode == x_diff:
            diff = x_difference(x_current, x_previous)
        
        elif mode == normalized_criterion:
            diff = norm(grad_function(x_current), ord=2) / norm(grad_function(x_0), ord=2)
        
        elif mode == x_opt_diff: 
            diff = norm(x_current - x_sol, ord=2)
        else: 
            # the last case is where the criterion is passed as an argument
            diff = mode(x_current)

        criterion_history.append(diff)
        # add 'x_current' to the history
        x_history.append(x_current)
        
        if diff <= eps: 
            break

    assert len(x_history) == k + 2, f"expected {k + 2} points. Found: {len(x_history)}"

    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

### __а). (5 points)__ Realize the Fletcher-Reeves and Polak-Ribier method. 

Describe how you will search for $\alpha_k$ steps (both the algorithm and its initialization are interesting). 

Add to the algorithms the ability to do "restarts" (sometimes take $\beta_k = 0$) with some frequency that can be customized.

1. The choice of the step length at any iteration $\alpha_k$ greatly contributes to the algorithm's robustness. 

* The initialization is as follows, at each iteration $\alpha_0 = \alpha_{k - 1} \cdot \frac{\nabla f_{k - 1} ^ T \cdot p_{k - 1}}{\nabla f_{k} ^ T \cdot p_k}$
where $\alpha_{k - 1}$ is the previous step size. The very first step size is generally $1$.

* The algorithm I used in based on the backtracking algorithm explained in the section $2$ of chapter $3$ in the [book](https://fmin.xyz/assets/files/NumericalOptimization.pdf) 
<!-- but with adding the 2nd Wolfe condition. The full version can be found through this link -->

* The linear search for the step size satisfies the wolph conditions  mentioned [here](https://en.wikipedia.org/wiki/Wolfe_conditions)

```python
def fr_beta(r_new: np.ndarray, r_old: np.ndarray):
    rn, ro = np.squeeze(r_new), np.squeeze(r_old)
    return (rn.dot(rn)) / (ro.dot(ro))

def pr_beta(r_new: np.ndarray, r_old: np.ndarray):
    r_new, r_old = np.squeeze(r_new), np.squeeze(r_old)
    return (r_new.dot(r_new - r_old)) / (r_old.dot(r_old))
```

```python
def descent(grad_function: callable, 
            x_k: np.ndarray, 
            p_k: np.ndarray):

    # make sure both x_k and p_k are column vectors
    p_k = np.expand_dims(p_k, axis=1) if p_k.ndim == 1 else p_k
    if p_k.ndim != 2: 
        raise ValueError(f"expected direction p_k with 2 dimensions. Found: {p_k.ndim} dimensions")

    d =  (grad_function(x_k).T @ p_k).item()
    assert isinstance(d, float), "The descent of a function in a given direction is scalar !!"
    return d

def wolfe_conditions(function: callable, 
                     grad_function: callable,
                     x_k: np.ndarray, 
                     p_k: np.ndarray, 
                     alpha: float, 
                     c1: float, 
                     c2: float, 
                     ):
    # calculate the next point given p_k, and alpha_k
    next_x = x_k + alpha * p_k

    d = descent(grad_function=grad_function, 
                x_k=x_k, 
                p_k=p_k)
    
    # the inequality 3.6a
    if function(x_k) + c1 * alpha * d < function(next_x):
        return False

    # inequality 3.6b
    return descent(grad_function=grad_function, x_k=next_x, p_k=p_k) >= c2 * d
```

```python
def step_size_search(function: callable, 
                     grad_function: callable,
                     x_k: np.ndarray, 
                     p_k: np.ndarray, 
                     c1: float = 10 ** -4,
                     c2: float = 0.9, 
                     search_coef: float = 0.8,
                     initial_alpha: float = 1,
                     min_alpha: float = 10 ** -3) -> float:

    # let's make sure the conditions of the algorithms are satisfied
    if search_coef < 0 or search_coef > 1 or initial_alpha < 0:
        raise ValueError((f"Please make sure the search coefficient is positive and less than 1"
                          f"\n and the initial alpha is positive. Found: alpha={initial_alpha} and coefficient: {search_coef}"))

    alpha = initial_alpha
    good_alpha = False    

    while alpha > min_alpha:
        good_alpha = wolfe_conditions(function=function, 
                                      grad_function=grad_function, 
                                      x_k=x_k,
                                      p_k=p_k,
                                      c1=c1,
                                      c2=c2, 
                                      alpha=alpha)
        if good_alpha: 
            break
    
        alpha = search_coef * alpha
    
    # if not good_alpha:
    #     print(f"None of the step sizes met the conditions. Returning the smallest value: {alpha}")
        
    return alpha 
```

```python
def conjugate_vector_method(function: callable, 
                      grad_function: callable, 
                      x_0: np.ndarray, 
                      beta: callable,
                      K = 10**3, 
                      eps = 10**-5, 
                      mode = Union[normalized_criterion, callable], 
                      return_history: bool = True,
                      restart_on:int=1,
                      x_sol: float = None,
                      c1: float = 10 ** -5, 
                      c2: float = 0.4
                      ): 

    # the first step is to make sure the 'mode' variable is defined above
    if isinstance(mode , str) and mode not in [f_diff, x_diff, normalized_criterion, x_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))
    
    if mode == x_opt_diff and x_sol is None: 
        raise ValueError(f"using mode = {x_opt_diff} requires passing the solution to the problem")
    
    x_current, x_previous = x_0.copy(), x_0.copy()
    r = -grad_function(x_current)
    r_old = r.copy()
    p_current, p_previous= r.copy(), r.copy()
    alpha = 1

    x_history = [x_current]
    criterion_history = []

    for k in tqdm(range(K)):
        # calculate the initial alpha to pass to the line search algorithm
        initial_alpha = 1 if k == 0 else (alpha * (grad_function(x_previous).T @ p_previous).item() / (grad_function(x_current).T @ p_current).item()) 

        # find alpha
        alpha = step_size_search(function=function, 
                                      grad_function=grad_function, 
                                      x_k=x_current, 
                                      p_k=p_current, 
                                      c1=c1, 
                                      initial_alpha=max(initial_alpha, 10),
                                      search_coef=(0.5 + 0.25 * (1 - 1 / ((k + 2)) ** 2)), 
                                      c2=c2
                                      )

        x_new = x_current + alpha * p_current
        r_new = -grad_function(x_new)

        if mode == f_diff:
            diff = f_difference(function(x_new), function(x_current))
        elif mode == x_diff:
            diff = x_difference(x_new, x_current)
        elif mode == normalized_criterion:
            diff = norm(grad_function(x_new), ord=2) / norm(grad_function(x_0), ord=2)
        elif mode == x_opt_diff: 
            try:
                diff = norm(x_new - x_sol, ord=2)
            except np.linalg.LinAlgError:
                diff = 10  # 10 should be cobsidered a very large error
        else: 
            # the last case is where the criterion is passed as an argument
            diff = mode(x_new)
        
        # add 'x_current' to the history
        x_history.append(x_new)
        # add the current criterion to history 
        criterion_history.append(diff)

        if diff <= eps:
            break

        if restart_on is not None and k % restart_on == 0:
            p_current = r_new
        else:
            beta_ = beta(r_new, r_old)
            p_current = r_new + beta_ * p_current
        
        r_old = r.copy()
        r = r_new.copy()
        x_previous = x_current.copy()
        x_current = x_new.copy()

    
    assert len(x_history) == k + 2, f"expected {k + 2} points. Found: {len(x_history)}"
    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

$$\min_{x \in \mathbb{R}^d} \left[\tfrac{1}{2} x^T A x - b x \right]$$
with a positive definite symmetric matrix $A \in \mathbb{R}^{d \times d}$ and some vector $b \in \mathbb{R}^d$. 

We need to learn how to generate a matrix $A$ with the possibility to specify its spectrum (eigenvalues). 

The previous assignment already asked us to do this. 

We suggest the following approach based on the decomposition $A = Q D Q^T$, where $D$ is a diagonal matrix formed from eigenvalues and $Q$ is orthogonal (it can be generated using the $QR$-decomposition of a random matrix).

Suppose we have a quadratic problem whose matrix $A \in \mathbb{R}^{d \times d}$ has clustered eigenvalues, meaning, that there exists some number of clusters $k \leq d$ and values $\tilde \lambda_1 < \ldots < \tilde \lambda_k$ such that for any $\lambda_i$ eigenvalue of matrix $A$ there exists $j \leq k$ such that $\lambda_i \in [(1 - p) \tilde \lambda_j; (1 + p) \tilde \lambda_j]$, where $p < 1$.

Then we will need to generate clustered eigenvalues and then the matrix $A$. When generating the spectrum, try to make sure that all values in it are different. 

As a convergence criterion, use $\frac{\| x^k - x^* \|^2_A} {\| x^0 - x^* \|^2_A}$, where $k$ is the iteration number and $\| x \|^2_A = \langle x, Ax \rangle$.

```python
def linear_function(x: np.ndarray, A: np.ndarray, b: np.ndarray):
    b = np.expand_dims(b, axis=-1) if b.ndim == 1 else b
    x = np.expand_dims(x, axis=-1) if x.ndim == 1 else x
    
    if A.shape[1] != x.shape[0]:
        raise ValueError(f"A: {A.shape}, x: {x.shape}") 

    if b.shape[0] != x.shape[0]:
        raise ValueError(f"b: {b.shape}, x: {x.shape}")

    res = A @ x - b

    assert res.shape == (A.shape[0], 1)

    return res
```

The conjugate method was initially introduced to solve the linear equation: 
$$ A \cdot x = b $$

The conjugate gradient method is more optimal for this case (in contrast with the non-linear extension of the algorithm)
 
I will implement the algorithm 5.2 in section $5$ from this [book](https://fmin.xyz/assets/files/NumericalOptimization.pdf)


```python
def linear_CG(A: np.ndarray,
              b: np.ndarray,
              x_0: np.ndarray, 
              eps: float = 10 ** -5, 
              mode: Union[str, callable] = x_opt_diff,
              return_history: bool = True, 
              K: int = 10 ** 3):

    # filter modes    
    if isinstance(mode , str) and mode != x_opt_diff:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))

    # extract the solution and convert it to expected shape
    x_sol = np.linalg.solve(A, b)
    x_sol = np.expand_dims(x_sol, axis=-1) if x_sol.ndim == 1 else x_sol
    
    # extra information to gurantee correctness 
    dim, _ = x_sol.shape    
    f = lambda x: linear_function(x, A, b)
    assert np.allclose(f(x_sol), 0), "make sure you are calling the solution correctly"
    
    # set the current x to the first one
    x_current = x_0
    
    r_current = f(x_0)
    r_previous = r_current.copy()
    
    p_current = -r_current.copy()

    alpha, beta = 0, 0

    x_history = [x_current]
    criterion_history = []

    for k in tqdm(range(K)):
        try: 
            alpha = (r_current.T @ r_current).item() / (p_current.T @ A @ p_current).item()

            x_current = x_current + alpha * p_current

            # set the current r to r_previous before updating r
            r_previous = r_current.copy()
            r_current = r_current + alpha * (A @ p_current)

            assert r_current.shape == (dim, 1)

            beta = (r_current.T @ r_current).item() / (r_previous.T @ r_previous).item()

            p_current = -r_current + beta * p_current   

            if mode == x_opt_diff: 
                diff = norm(x_current - x_sol, ord=2)
            else: 
                # the last case is where the criterion is passed as an argument
                diff = mode(x_current)

            criterion_history.append(diff)
            # add 'x_current' to the history
            x_history.append(x_current)
        
            if diff <= eps:
                break
        except ZeroDivisionError:
            # this probably means the method converged and values are set to 0 due to numerical precisions
            break 
    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]

```

```python
from scipy.linalg import orth

def eig_matrix_generation(
    dim: int,
    clusters: List[Tuple[int, int]],
    p: float,
):
    # make sure the sizes of clusters sum up to the number of dimensions
    assert sum([x for x, _ in clusters]) == dim, f"Please make sure to produce {dim} eigenvalues !!"
    
    eigs = []
    for num_eig, center in clusters:
        eigs.extend(((np.random.rand(num_eig) - .5) * 2 * p + center).tolist())
    
    # generate an orthogonal matrix Q
    Q = orth(np.random.rand(dim, dim))

    # Create the diagonal matrix D from eigenvalues
    D = np.diag(np.array(eigs))

    # generate the final matrix
    A = Q @ D @ Q.T

    assert A.shape == (dim, dim)
    
    return A

matrices = [
    eig_matrix_generation(100, [(50, 1), (50, 10)], 0.05),
    eig_matrix_generation(100, [(50, 1), (50, 100)], 0.05),
    eig_matrix_generation(100, [(50, 1), (50, 1000)], 0.05),
    eig_matrix_generation(100, [(50, 1), (50, 10000)], 0.05),
    eig_matrix_generation(100, [(50, 1), (50, 100000)], 0.05),
]
```

```python
def newton_method( function: callable, 
                    grad_function: callable, 
                    hessian_function: callable,
                    x_0: np.ndarray, 
                    gamma_k: callable = None,
                    K = 10**3, 
                    eps = 10**-5, 
                    mode = normalized_criterion, 
                    return_history: bool = True):
    # the default step size is '1'
    gamma_k = (lambda _ : 1) if gamma_k is None else gamma_k
    
    if mode not in [f_diff, x_diff, normalized_criterion]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion]}"
                          f"Found: {mode}"))
    # a boolean flag to determine whether to stop the iterative process
    criterion_stop = False
    # counter for the number of iterations
    k = 0
    # set the variables for the parameters 'x', 'y'
    x_current, x_previous = x_0, x_0

    # create a list to save the intermediate points  
    x_history = [x_current]
    criterion_history = [(1 if mode==normalized_criterion else 0)]

    while not criterion_stop:
        x_previous = x_current
        gamma = gamma_k(k)
        current_hessian = hessian_function(x_current)
        # if the current hessian contains 'nan' values
        if np.sum(np.isnan(current_hessian)) > 0:
            raise ValueError(f"the hessian contains nan values")
        
        # carry out with the inverse operation
        inv_hessian = np.linalg.pinv(current_hessian)
        if np.sum(np.isnan(inv_hessian)) > 0:
            raise ValueError(f"The inverse of the hessian matrix conains ")

        # make sure to use @: the matrix multiplication operator and not '*', the element-wise multiplication
        x_current =  x_current - gamma * inv_hessian @ grad_function(x_current) 

        if mode == f_diff:
            diff = f_difference(function(x_current), function(x_current))
        elif mode == x_difference:
            diff = x_difference(x_current, x_previous)
        else:
            diff = norm(grad_function(x_current), ord=2) / norm(grad_function(x_0), ord=2)

        criterion_history.append(diff)
        # add 'x_current' to the history
        x_history.append(x_current)
        # increment 'k' 
        k += 1
        criterion_stop = diff <= eps or k >= K
    
    assert len(x_history) == k + 1, f"expected {k + 1} points. Found: {len(x_history)}"
    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

### __b). (7 points)__ 
Implement the quasi-Newton method BFGS. 

Use it to solve the regression problem.

The implementation below is mainly based on the algorithm description from this [link](https://en.wikipedia.org/wiki/Broyden%E2%80%93Fletcher%E2%80%93Goldfarb%E2%80%93Shanno_algorithm)

The linear search for the step size satisfies the wolph conditions  mentioned [here](https://en.wikipedia.org/wiki/Wolfe_conditions)

```python
def descent(grad_function: callable, 
            x_k: np.ndarray, 
            p_k: np.ndarray):

    # make sure both x_k and p_k are column vectors
    p_k = np.expand_dims(p_k, axis=1) if p_k.ndim == 1 else p_k
    if p_k.ndim != 2: 
        raise ValueError(f"expected direction p_k with 2 dimensions. Found: {p_k.ndim} dimensions")

    d =  (grad_function(x_k).T @ p_k).item()
    assert isinstance(d, float), "The descent of a function in a given direction is scalar !!"
    return d

def wolfe_conditions(function: callable, 
                     grad_function: callable,
                     x_k: np.ndarray, 
                     p_k: np.ndarray, 
                     alpha: float, 
                     c1: float, 
                     c2: float, 
                     ):
    # calculate the next point given p_k, and alpha_k
    next_x = x_k + alpha * p_k

    d = descent(grad_function=grad_function, 
                x_k=x_k, 
                p_k=p_k)
    
    # the inequality 3.6a
    if function(x_k) + c1 * alpha * d < function(next_x):
        return False

    # inequality 3.6b
    return descent(grad_function=grad_function, x_k=next_x, p_k=p_k) >= c2 * d
```

The implementation of the linear search algorithm is based on the Backtracking Line Search algo explained in the [book](https://fmin.xyz/assets/files/NumericalOptimization.pdf)

```python
def step_size_search(function: callable, 
                     grad_function: callable,
                     x_k: np.ndarray, 
                     p_k: np.ndarray, 
                     c1: float = 10 ** -4,
                     c2: float = 0.9, 
                     search_coef: float = 0.8,
                     initial_alpha: float = 1,
                     max_iterations: int = 10) -> float:

    # let's make sure the conditions of the algorithms are satisfied
    if search_coef < 0 or search_coef > 1 or initial_alpha < 0:
        raise ValueError((f"Please make sure the search coefficient is positive and less than 1"
                          f"\n and the initial alpha is positive. Found: alpha={initial_alpha} and coefficient: {search_coef}"))

    alpha = initial_alpha
    good_alpha = False    

    for _ in range(max_iterations):
        good_alpha = wolfe_conditions(function=function, 
                                      grad_function=grad_function, 
                                      x_k=x_k,
                                      p_k=p_k,
                                      c1=c1,
                                      c2=c2, 
                                      alpha=alpha)
        if good_alpha: 
            break
    
        alpha = search_coef * alpha
    
    if not good_alpha:
        print(f"None of the step sizes met the conditions. Returning the smallest value: {alpha}")
        
    return alpha 
   
```

```python
def bfgs(function: callable,
         grad_function: callable,
         x_0: np.ndarray,
         B_0: np.ndarray = None,
         eps: int = 10 ** -5,
         K: int = 10 ** 3,
         mode = normalized_criterion,
         return_history:bool=True,

         c1: float = 10 ** -4, 
         c2: float = 0.9,
         search_coef: float = 0.8, 
         initial_step_size: float = 0.8, 
         max_search_iterations: int = 10):
    
     # the default value for the first B matrix is the identity matrix 
     B_0 = (B_0) if (B_0 is not None) else np.eye(max(x_0.shape))
    
     # the first step is to make sure the 'mode' variable is defined above
     if mode not in [f_diff, x_diff, normalized_criterion]:
          raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion]}"
          f"Found: {mode}"))
    
     # a boolean flag to determine whether to stop the iterative process
     criterion_stop = False
     k = 0
     # set the variables for the parameters 'x', 'y'
     x_current, x_previous = x_0, x_0
     B_k = B_0 # B_k doesn't represent the estimation of the hessian but the estimation of its inverse. 

     # create a list to save the intermediate points  
     x_history = [x_current]
     criterion_history = []
     alpha_k = None

     while not criterion_stop:
          # start with setting x_previous
          x_previous = x_current

          # calculate p_k: step 1
          x_grad = grad_function(x_current)
          # define p_k as - grad * B
          p_k = - B_k @ x_grad

          if p_k.shape != x_current.shape:
               raise ValueError(f"x: {x_current.shape}, p_k: {p_k.shape}")

          # find the step_size: step 2
          alpha_k = step_size_search(function=function,
                                   grad_function=grad_function,
                                   x_k=x_current, 
                                   p_k=p_k,
                                   c1=c1, 
                                   c2=c2,
                                   search_coef=search_coef, 
                                   initial_alpha=(alpha_k if alpha_k is not None else initial_step_size),
                                   max_iterations=max_search_iterations,
                                   )
          # calculate s_k and x_k+1: step 3
          s_k = alpha_k * p_k 
          x_current = x_current + s_k
          
          # calculate y_k: step4:
          y_k = grad_function(x_current) - grad_function(x_previous)

          # step 5: update the inverse of the marix: 
          # I am using the closed formula
          sy = (s_k.T @ y_k).item()

          # before updating the hessian matrix approximation
          # let's make sure the shapes are correct
          if x_current.ndim != 2 or s_k.shape != (x_current).shape or y_k.shape != (x_current).shape:
               raise ValueError(f"s_k: {s_k.shape}, y_k: {y_k.shape}")

          B_k = B_k + ((sy + (y_k.T @ B_k @ y_k).item()) * s_k @ s_k.T) / (sy ** 2) - (B_k @ y_k @ s_k.T + s_k @ y_k.T @ B_k) / (sy)

          if mode == f_diff:
               diff = f_difference(function(x_current), function(x_current))
          elif mode == x_difference:
               diff = x_difference(x_current, x_previous)
          else:
            diff = norm(grad_function(x_current), ord=2) / norm(grad_function(x_0), ord=2)

          criterion_history.append(diff)
          # add 'x_current' to the history
          x_history.append(x_current)
          # increment 'k' 
          k += 1

          # stop the iterative process either because the maximum number of iterations is reached 
          # or the difference is less than epsilon
          criterion_stop = diff <= eps or k >= K

     assert len(x_history) == k + 1

     # the function will all intermediate points if the 'return_history' argument is set to True
     # and only the last point otherwise    
     return (x_history, criterion_history) if return_history else x_history[-1]
```

```python
x_opt_diff= '|x * - x_k|'

import warnings 
# let's start with a scalar implementation of newton method 
def scalar_newton_method(function: callable, 
                    d1_function: callable, 
                    d2_function: callable,
                    x_0: float, 
                    gamma_k: callable = None,
                    K: int=10**3, 
                    eps:float=10**-5, 
                    mode:str=normalized_criterion, 
                    return_history: bool = True,
                    x_sol: float = None):
    # the default step size is '1'
    gamma_k = (lambda _ : 1) if gamma_k is None else gamma_k
    
    if mode not in [f_diff, x_diff, normalized_criterion, x_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))
    
    if mode == x_opt_diff and x_sol is None: 
        raise ValueError(f"using mode = {x_opt_diff} requires passing the solution to the problem")
    
    # a boolean flag to determine whether to stop the iterative process
    criterion_stop = False
    # counter for the number of iterations
    k = 0
    # set the variables for the parameters 'x', 'y'
    x_current, x_previous = x_0, x_0
    # create a list to save the intermediate points  
    x_history = [x_current]
    criterion_history = []

    while not criterion_stop:
        x_previous = x_current
        gamma = gamma_k(k)
        # compute the first and 2nd derivative of 'f'
        d1_current, d2_current = d1_function(x_current), d2_function(x_current) 
        
        # make sure the d2_current is positive
        if d2_current <= 0:
            warnings.warn(f"The 2nd derivative is expected to be positive. Found: {d2_current}")

        x_current = x_current - gamma * (d1_current / (d2_current + np.finfo(np.float64).eps))
        
        if mode == f_diff:
            diff = f_difference(function(x_current), function(x_current))
        elif mode == x_difference:
            diff = x_difference(x_current, x_previous)
        elif mode == normalized_criterion:
            diff = abs(d1_current) / abs(d1_function(x_0))
        else:
            diff = abs(x_sol - x_current)

        criterion_history.append(diff)
        # add 'x_current' to the history
        x_history.append(x_current)
        # increment 'k' 
        k += 1
        criterion_stop = (diff <= eps) or (k >= K)
    
    assert len(x_history) == k + 1, f"expected {k + 1} points. Found: {len(x_history)}"
    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

#### Newton Method with Cubic Regularization
Before diving into the implementation, let's go through the theoretical part. 

The algorithm is based on the following assumption, for a set $F \in \mathbb{R}^n$ where $n$ is the problem's dimension, the hessian matrix of $f$ is $L$-Lipschitz. The algorithm also defines the following: 

$$ f_{M,x}(y) = < f^{'} (x), y - x > + \frac{1}{2} < f^{''}(x) (y - x), y -x>  + \frac{M}{6} \cdot \| x - y \| ^ 3$$

defining 
$$T_{M}(x) = \argmin_{y \in \mathbb{R}^n} f_{M, x}(y)$$
and 
$$\bar{f}_{M}(x) = \min_{y \in \mathbb{R}^n} f_{M,x}(y)$$

The algorithm is as follows: 

1. initialization: $x_0 \in F$, $L_0 < L$
2. Find $M_k \in [L_0, L]$ such that $$f(T_{M_k}(x_k)) \leq \bar{f}_{M_k}(x_k)$$
3. Set the $x_{k + 1} = T_{M_k}(x_k)$

The authors suggest the following approach to find $M_k$
1. The authors established theoretically that: 
    $$ f(T_{M_k}(x_k)) \leq \bar{f}_{M_k}(x_k), ~~ \forall M_k \geq L$$
2. while $f(T_{M_k}(x_k)) > \bar{f}_{M_k}(x_k)$, $M_K := 2 \cdot M_k$
3. $x_{k + 1} := T_{M_k}(x_k)$ and $M_{k + 1} := \max(\frac{1}{2} \cdot M_k, L_0) $

Let's find the value of $L$ for our one-dimensional problem. 

We have
\begin{align}
|f^{''}(x) - f^{''}(y)| &= |\frac{1}{1 + x^2} - \frac{1}{1 + y^2}| \\
|f^{''}(x) - f^{''}(y)| &= |\frac{x^2 - y^2}{(1+x^2) \cdot (1 + y^2)}| \\
\end{align}

$L$-Lispschitz condition is equivalent to
\begin{align}
|f^{''}(x) - f^{''}(y)| &\leq L |x - y| \\
\iff |\frac{x^2 - y^2}{(1+x^2) \cdot (1 + y^2)}| &\leq L |x - y| \\
\iff |\frac{x + y}{(1+x^2) \cdot (1 + y^2)}| &\leq L && \text{since the equation is satisfied $\forall L, x = y$} \\
\end{align}

We can estimate $L$ as follows

\begin{align}
|\frac{x + y}{(1+x^2) \cdot (1 + y^2)}| &\leq \frac{|x| + |y|}{(1+x^2) \cdot (1 + y^2)}  = \frac{|x|}{(1+ x^2) \cdot (1 + y^2)} + \frac{|y|}{(1 + x^2)(1 + y^2)}&& \text{using $|x + y| \leq |x| + |y|$} \\
|\frac{x + y}{(1+x^2) \cdot (1 + y^2)}| &\leq \frac{1}{2(1 + y^2)} + \frac{1}{2(1 + x^2)}  && \text{using $1 + x^2 \geq 2|x|$ and $1 + x^2 \geq 1$}\\
|\frac{x + y}{(1+x^2) \cdot (1 + y^2)}| &\leq 1  
\end{align}

Then $L = 1$ for our problem.

```python
# I will use the gradient descent to optimize the sub problem in the cubic newton method
def scalar_gradient_descent(function: callable, 
                     d1_function: callable, 
                     x_0: np.ndarray,                      
                     K: int = 10 ** 3,
                     eps: float = 10 ** -5, 
                     mode: str = normalized_criterion,
                     gamma_k: callable = None,                      
                     return_history: bool = False,
                     x_sol: float = None
                     ) -> Union[List[np.ndarray], np.ndarray]:

    if mode not in [f_diff, x_diff, normalized_criterion, x_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))
    
    if mode == x_opt_diff and x_sol is None: 
        raise ValueError(f"using mode = {x_opt_diff} requires passing the solution to the problem")

    # a boolean flag to determine whether to stop the iterative process
    criterion_stop = False
    # counter for the number of iterations
    k = 0
    x_current, x_previous = x_0, x_0
    # create a list to save the intermediate points  
    x_history = [x_current]
    criterion_history = []

    while not criterion_stop:
        # save the k-th point
        x_previous = x_current

        gamma = gamma_k(k)

        d1_current = d1_function(x_current)

        # compute the (k+1)-th point
        x_current = x_current - gamma * d1_current  

        # compute the criterion
        if mode == f_diff:
            diff = f_difference(function(x_current), function(x_current))
        elif mode == x_difference:
            diff = x_difference(x_current, x_previous)
        elif mode == normalized_criterion:
            diff = abs(d1_function(x_current)) / abs(d1_function(x_0))
        else:
            diff = abs(x_sol - x_current)

        criterion_history.append(diff)
        # add 'x_current' to the history
        x_history.append(x_current)
        # increment 'k' 
        k += 1
        # stop the iterative process either because the maximum number of iterations is reached 
        # or the difference is less than epsilon
        criterion_stop = diff <= eps or k >= K
    
    assert len(x_history) == k + 1
    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

```python
import random
import math

def cubic_newton_setup(L: float = 1, x_0: float = None, L_0:float=None, seed:int=69):
    random.seed(seed)
    np.random.seed(seed)
    if L_0 is not None and L_0 >= L:
        raise ValueError(f"The L_0 constant is expected to be smaller than 'L'. Found: L_0: {L_0} and L: {L}")
    
    # L / L_0 should be large to have the opportunity to experiment with smaller values of M (otherwise M will reach 'L' quite fast)
    L_0 = L / 1000 + 0.01 * random.random()
    x_0 = (5 * random.random()) if x_0 is None else x_0
    M_0 = L_0 + 0.5
    return L, L_0, x_0, M_0

def cubic_approximation_Mx(y: float,
                        function: callable, 
                        d1_function: callable,
                        d2_function: callable, 
                        x_current: float,
                        M: float):
    
    t1 = function(x_current)
    t2 = d1_function(x_current) * (y - x_current)
    t3 = 0.5 * d2_function(x_current) * (y - x_current) ** 2
    t4 = (M / 6) * abs(y - x_current) ** 3

    return t1 + t2 + t3 + t4

def derivative_cubic_approximation_Mx(y: float, 
                                      function: callable, 
                                      d1_function: callable, 
                                      d2_function: callable, 
                                      x_current:float, 
                                      M:float):
    t1 = d1_function(x_current)
    t2 = d2_function(x_current) * (y - x_current)
    t3 = 0.5 * M * abs(y - x_current) * (y - x_current)
    return t1 + t2 + t3
```

```python
# let's define the update process
def cubic_newton_update(function: callable, 
           d1_function: callable, 
           d2_function: callable, 
           x_current: float, 
           M_current: float, 
           L:float,
           L_0: float,
           gd_k: int = 10 ** 3,
           ) -> Tuple[float, float]:
    M = M_current

    while True:
        # first define the model approximation for the given M
        cubic_approx = lambda y: cubic_approximation_Mx(y=y, 
                                                        function=function, 
                                                        d1_function=d1_function, 
                                                        d2_function=d2_function, 
                                                        x_current=x_current, 
                                                        M=M)
        # define its derivative
        cubic_approx_derivative = lambda y: derivative_cubic_approximation_Mx(y=y, 
                                                                                function=function, 
                                                                                d1_function=d1_function, 
                                                                                d2_function=d2_function,
                                                                                x_current=x_current, 
                                                                                M=M)
        # solve the sub problem
        x_points, _ = scalar_gradient_descent(function=cubic_approx, 
                                                    d1_function=cubic_approx_derivative,
                                                    x_0=x_current, 
                                                    return_history=True,
                                                    K=gd_k,
                                                    gamma_k=lambda k: 1 / L)

        # iterate through the last 10 points found by gradient descent
        candidates = x_points[-10:][::-1]
        for xp in candidates:
            # the first term is the value of the function at the approximate solution
            t1 = function(xp) 
            t2 = cubic_approx(y=xp)
            if t1 <= t2: 
                # time to break the look and return x_p as the next value of 'x' and M as the next value of M
                return xp, max(M / 2, L_0)

        # at this point, 
        # we know we do not have a good solution, we need to update m
        
        if M > 2 * L:
            raise ValueError((f"M exceeded 2 * L, Please make sure to verify either the approximation, " 
                             f"its gradient or increase the number of iterations"))
        M = 2 * M
```

```python
def cubic_newton_method(function: callable, 
                        d1_function: callable, 
                        d2_function: callable,
                        L:float, 
                        K: int = 10 ** 3,
                        eps: float = 10 ** -5, 
                        mode: str = normalized_criterion,
                        return_history: bool = False,
                        x_sol: float = None,
                        x_0: float = None,
                        seed: int = 69,
                        ) -> Union[List[np.ndarray], np.ndarray]:

    if mode not in [f_diff, x_diff, normalized_criterion, x_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[f_diff, x_diff, normalized_criterion, x_opt_diff]}"
                          f"Found: {mode}"))
    
    if mode == x_opt_diff and x_sol is None: 
        raise ValueError(f"using mode = {x_opt_diff} requires passing the solution to the problem")

    # set up the initial values
    L, L_0, x_0, M_0 = cubic_newton_setup(L=L,x_0=x_0, seed=seed)
    M_current = M_0
    x_current = x_0
    criterion_stop = False
    k = 0
    x_history = [x_current]
    criterion_history = []

    while not criterion_stop:
        # save the k-th point
        x_previous = x_current
        # perform the update step
        x_current, M_current = cubic_newton_update(function=function,
                                                   d1_function=d1_function,
                                                   d2_function=d2_function,
                                                   x_current=x_current, 
                                                   M_current=M_current, 
                                                   L=L, 
                                                   L_0=L_0
                                                   )

        # compute the criterion
        if mode == f_diff:
            diff = f_difference(function(x_current), function(x_current))
        elif mode == x_difference:
            diff = x_difference(x_current, x_previous)
        elif mode == normalized_criterion:
            diff = abs(d1_function(x_current)) / abs(d1_function(x_0))
        else:
            diff = abs(x_sol - x_current)

        criterion_history.append(diff)
        # add 'x_current' to the history
        x_history.append(x_current)
        # increment 'k' 
        k += 1
        # stop the iterative process either because the maximum number of iterations is reached 
        # or the difference is less than epsilon
        criterion_stop = diff <= eps or k >= K
    
    assert len(x_history) == k + 1
    # the function will all intermediate points if the 'return_history' argument is set to True
    # and only the last point otherwise    
    return (x_history, criterion_history) if return_history else x_history[-1]
```

The code below can be considered the 'training' phase of the linear model. Since the labels are either (-1, 1), we should use the tanh activation layer:

1. first calculate the logits: 
$$l_i =  \tanh(w_i ^ T \cdot x_i) $$ 
2. $$\hat{y_i} =  \begin{cases} -1 ~ l_i < 0 \\ 1 ~~ \text{otherwise}\end{cases}$$

```python
def predict(w: np.ndarray, X: np.ndarray = None):
    X = X_test if X is not None else X
    l = np.tanh(X @ w)
    preds = -1 + 2 * (l >= 0)
    return preds


def accuracy(y_pred, y_true):

    assert y_pred.shape == y_true.shape, f"preds: {y_pred.shape}, labels: {y_true.shape}"
    return np.mean(y_pred == y_true)
```
