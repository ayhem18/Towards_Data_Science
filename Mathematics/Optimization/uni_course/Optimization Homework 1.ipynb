Let us start with the definition of a convexity of a set: 

A set $Q$ is said to be convex over $\mathbb{R} ^ {d} $ if: 

$\forall x \in Q , y \in Q, \forall \alpha \in (0, 1)$, 

$\alpha \cdot x + (1 - \alpha) \cdot y \in Q $

$x \in S$ and $y \in S$, we have $x_1, x_2 > 0$ and $y_1, y_2 > 0$ and since $\alpha, 1 - \alpha > 0$, then any linear combination is positive. Therefore,
1. $z_1 > 0, z_2 > 0$

* method:

Let's consider the following procedure: 
1. Given, $d$, $L$, $\mu$ we can create a diagonal matrix of dimension $d$ where all values are equally spaced in the range $[\mu, L]$
2. Generate a orthonormal (orthogonal + with norm 1) matrix $Q$ and assign $A$ as $A = Q \cdot P \cdot Q^T$

* Proof:

1. Let's prove the function is positive semi definite:

since each entry in $P$ is positive, we can write $P = P_{r}^ 2$ where $P_{r}$ is a diagonal matrix with entries equal to the square root of the entries in $P$. Then, we have 
\begin{align}
A &= Q \cdot P_{r} \cdot P_{r} \cdot Q^{T} \\
A &= Q \cdot P_{r} \cdot P_{r}^T \cdot Q^{T} \\ 
A &= R \cdot R^{T} && \text{ where $R = Q \cdot P_r$}
\end{align}
matrices of such form are known to be positive semi definite

2. Let's prove the eigenvalues belong to the mentioned range.

Orthonormal matrices satisfy the property $Q^T = Q^{-1}$. Assuming 
$$
Q = 
\begin{bmatrix} 
q_1 && q_2 && .. && q_d
\end{bmatrix}
$$
since $Q^T \cdot Q = I$, then $Q^T \cdot q_i = 1_{k == i}$ (vector with zeros everywhere except the $i$-th entry and $1$ in the $i$-th entry) 
\begin{align}
Q \cdot P \cdot Q^T \cdot q_i =  Q \cdot P \cdot  1_{k == i} \\
Q \cdot P \cdot Q^T \cdot q_i =  P_i \cdot q_i 
\end{align}

The latter proves that the entries of the matrix $P$ are eigenvalues of $A$.

The suggested method guarantees the problem is $L$ smooth and $\mu$ strong-convex

```python
import numpy as np
from scipy.stats import ortho_group

def generate_matrix_A(dim: int, mu: float, L:float) -> np.ndarray:
    # generate the orthonormal matrix
    q = ortho_group.rvs(dim)
    eigen = np.linspace(start=mu, stop=L, num=dim)

    # generate the diagonal matrix
    d = np.eye(dim)
    
    # set the eigenvalues in the diagonal entries
    for i, e in enumerate(eigen):
        d[i][i] = e

    if L <= mu  / 2:
        raise ValueError(f"The function expects 'mu' to be less than 2 * 'L'")
    
    return q @ d @ np.transpose(q)

a = generate_matrix_A(dim=5, mu=0.05, L= 0.1)
```

__d). (5 points)__  Implement the gradient descent method:
\begin{equation}
x_{k+1} = x_k - \gamma_k \nabla f(x_k),
\end{equation}
for a quadratic function with the matrix generated above and an arbitrary vector $b$ (preferably non-zero). It is already desirable to implement both the method and the gradient calculation as separate functions, with the method taking the target function itself (or its gradient) as a parameter.
```python
# let's write the modes independently 
from numpy.linalg import norm

x_opt_diff = 'x_k - x^*'
f_opt_diff = 'f(x_k) - f(x^*)'
x_diff = 'x_k+1 - x_k'
f_diff = 'f(x_k+1) - f(x_k)'


def x_difference(x1: np.ndarray, x2: np.ndarray) -> float:
    return norm(x1 - x2, ord=2)

def f_difference(f1: float, f2: float) -> float:
    # the expression f_x_k - f_sol is equivalent since the problem is minimization,
    # but the 'abs' function was used to make the method general and not only specific to the given problem
    return abs(f1 - f2)  
```

```python
def gradient_descent(function: callable, 
                     grad_function: callable, 
                     x_0: np.ndarray, 
                     x_sol: np.ndarray,
                     gamma_k: callable, 
                     K: int = 10 ** 3,
                     eps: float = 10 ** -5, 
                     mode: str = x_opt_diff,
                     return_history: bool = False):
    
    # the first step is to make sure the 'mode' variable is defined above
    if mode not in [x_opt_diff, f_opt_diff, x_diff, f_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[x_opt_diff, f_opt_diff, x_diff, f_opt_diff]}"
                          f"Found: {mode}"))
    # find the minimum value of the function   
    f_sol = function(x_sol)

    # a boolean variable to determine whether to stop the iterative process
    criterion_stop = False
    # counter for the number of iterations
    k = 0
    x_current, x_previous = x_0, x_0

    # create a list to save the intermediate points  
    x_history = [x_current]

    while not criterion_stop:
        # save the current point
        x_previous = x_current
        # update the point
        x_current = x_current - gamma_k(k) * grad_function(x_current)  

        # check the stopping criterion
        if mode == x_opt_diff:
            diff = x_difference(x_sol, x_current)
        elif mode == x_diff:
            diff = x_difference(x_previous, x_current)
        
        elif mode == f_diff:
            diff = f_difference(function(x_previous), function(x_current))
        else:
            diff = f_difference(function(x_current), f_sol)

        # add 'x_current' to the history
        x_history.append(x_current)

        # increment 'k' 
        k += 1

        # stop the iterative process either because the maximum number of iterations is reached 
        # or the difference is less than epsilon
        criterion_stop = diff <= eps or k > K
    
    assert len(x_history) == k + 1
    
    # the fucntion should return the 'numerical solution': x_current
    return x_history if return_history else x_history[-1]
```
