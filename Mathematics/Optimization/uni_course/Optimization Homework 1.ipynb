A set $Q$ is said to be convex over $\mathbb{R} ^ {d} $ if: 

$\forall x \in Q , y \in Q, \forall \alpha \in (0, 1)$,

## A)
__а). (4 points)__ Is the given problem always convex/strongly convex? Under what conditions?

* convexity:

For a differetional function to be convex, the hessian matrix $\nabla^2 f$ must be positive semi definite (psd). 

As for our problem we have: 
\begin{align}
\nabla ^ 2 f = \frac{1}{2} (A^T + A) ~ ~ psd & \iff  x^T \cdot (A + A^T) x \geq 0 ~~ \forall x \in \mathbb{R}^d \\
& \iff x^T \cdot A \cdot x + x^T \cdot A^T \cdot x \geq 0 \\
x^T \cdot A \cdot x &= (x^T \cdot A \cdot x) ^T && \text{$x^T \cdot A \cdot x$ is a scalar} \\
\implies x^T \cdot A \cdot x &=  x^T \cdot A^T \cdot x \\
\implies x^T \cdot A \cdot x + x^T \cdot A^T \cdot x \geq 0 & \iff 2 \cdot (x^T \cdot A \cdot x) \\
\nabla ^ 2 f ~~ psd & \iff A ~ positive ~ semi ~ definite  
\end{align}
Thus the condition for convexity of the problem is
$$ 
A ~ positive ~ semi ~ definite
$$
since $A \in \mathbb{R}^d$ , then: 

$$ A ~~ psd \implies eig(A) \geq 0$$

* strong convexity

For a problem to be strongly convex with constant $\mu$, we need:
\begin{align}
\nabla ^ 2 f - \mu \cdot I_d ~ ~ psd & \iff  \frac{1}{2} (A^T + A) - \mu \cdot I_d ~ ~ psd \\
& \iff  x^T \cdot (\frac{1}{2} (A + A^T) - \mu \cdot I_d ) x \geq 0 ~ ~ ~ \forall x \in \mathbb{R}^d \\
& \iff  x^T \cdot (\frac{1}{2} (A - \mu \cdot I_d) + \frac{1}{2} (A^T - \mu \cdot I_d) ) x \geq 0  \\
\end{align}
Using the fact that $(A - \mu \cdot I_d)^T = (A^T - \mu \cdot I_d)$, the last inequality is equivalent to:

$$ (A - \mu \cdot I_d) ~ ~ psd$$

\begin{align*}
(A - \mu \cdot I_d) ~ ~ psd & \implies eig(A - \mu \cdot I_d) \geq 0 \\
& \implies  eig(A) \geq \mu
\end{align*}

## C)
__c). (4 points)__ Propose a way to generate matrix $A$ so that $f$ is strongly convex and it is possible to set and change $d, \mu, L$. Prove that the method is correct. The diagonal matrix will give only 1 point.

We proved that for the problem to be $L$ smooth and $\mu$ strongly convex, 2 conditions must be satisfied:

Let's consider the case where $A$ is symmetric, the first condition is left unchanged, but:

\begin{align}
\frac{1}{2} \|(A + A^T)\|_2 & \leq L \\
\frac{1}{2} 2 \cdot \|A \|_2 & \leq L \\

\end{align}

Additionally if $A$ is symmetric, we have the following equivalence:

$$ A = A^T, eig(A) \geq 0 \iff A ~~ psd$$ 

which means that generating a symmetric matrix such that:
$$ \mu \leq \lambda(A) \leq L$$ 
guarantees that the problem is $\mu$ strongly convex and $L$ Lipschitz 

* method:

In the previous question we proved that for matrices $A$ such that:

1. all eigenvalues of $A$ lie in the range $[ \mu, L]$ 
2. Positive Semi Definite

The problem is strongly convex with constant $\mu$ and $L$ smooth.

Let's consider the following procedure: 
1. Given, $d$, $L$, $\mu$ we can create a diagonal matrix of dimension $d$ where all values are equally spaced in the range $[\mu, L]$
2. Generate a orthonormal (orthogonal + with columns of norm 1) matrix $Q$ and assign $A$ as $A = Q \cdot P \cdot Q^T$

* Proof:

1. Let's prove the function is positive semi definite:

since each entry in $P$ is positive, we can write $P = P_{r}^ 2$ where $P_{r}$ is a diagonal matrix with entries equal to the square root of the entries in $P$. Then, we have 
\begin{align}
A &= Q \cdot P_{r} \cdot P_{r} \cdot Q^{T} \\
A &= Q \cdot P_{r} \cdot P_{r}^T \cdot Q^{T} \\ 
A &= R \cdot R^{T} && \text{ where $R = Q \cdot P_r$}
\end{align}
matrices of such form are known to be positive semi definite

2. Let's prove the eigenvalues belong to the mentioned range.

Orthonormal matrices satisfy the property $Q^T = Q^{-1}$. Assuming 
$$
Q = 
\begin{bmatrix} 
q_1 && q_2 && .. && q_d
\end{bmatrix}
$$
since $Q^T \cdot Q = I$, then $Q^T \cdot q_i = 1_{k == i}$ (vector with zeros everywhere except the $i$-th entry and $1$ in the $i$-th entry) 
\begin{align}
Q \cdot P \cdot Q^T \cdot q_i =  Q \cdot P \cdot  1_{k == i} \\
Q \cdot P \cdot Q^T \cdot q_i =  P_i \cdot q_i 
\end{align}

The latter proves that the entries of the matrix $P$ are eigenvalues of $A$.

The suggested method guarantees the problem is $L$ smooth and $\mu$ strong-convex

```python
from typing import Iterable

import numpy as np
from scipy.stats import ortho_group

def generate_matrix_A(dim: int, mu: float, L:float) -> np.ndarray:
    # generate the orthonormal matrix
    q = ortho_group.rvs(dim)
    eigen = np.linspace(start=mu, stop=L, num=dim)

    assert len(eigen) == dim
    
    # generate the diagonal matrix
    d = np.eye(dim)
    
    # set the eigenvalues in the diagonal entries
    for i, e in enumerate(eigen):
        d[i][i] = e

    if L <= mu:
        raise ValueError(f"The function expects 'mu' to be less than 2 * 'L'")
    
    return q @ d @ np.transpose(q)

a = generate_matrix_A(dim=5, mu=0.05, L= 0.1)
```

```python
# ket's write the modes as fixed strings to avoid unexpected bugs 
from numpy.linalg import norm

x_opt_diff = 'x_k - x^*'
f_opt_diff = 'f(x_k) - f(x^*)'
x_diff = 'x_k+1 - x_k'
f_diff = 'f(x_k+1) - f(x_k)'

def x_difference(x1: np.ndarray, x2: np.ndarray) -> float:
    """Computes the euclidean norm of the difference vector 

    Args:
        x1 (np.ndarray): first vector
        x2 (np.ndarray): second vector

    Returns:
        float: returns the norm of the difference
    """
    return norm(x1 - x2, ord=2)

def f_difference(f1: float, f2: float) -> float:
    """returns the absolute difference between 2 values """
    # the expression f_x_k - f_sol is equivalent since the problem is minimization,
    # but the 'abs' function was used to make the method general and not only specific to the given problem
    return abs(f1 - f2)  
```

```python
def greedy_step(x: np.ndarray, 
                matrix_A: np.ndarray, 
                vector_B: np.ndarray, 
                grad_function: callable) -> float:
    """This function calculate the best step size (leading to the lowest value of f from the current point)
    The formula is written as answer to question 'H.2'
     
    Args:
        x (np.ndarray): the current point
        matrix_A (np.ndarray): the matrix A in the problem
        vector_B (np.ndarray): the vector B in the problem
        grad_function (callable): a function that returns the value of the gradient of a given point

    Returns:
        float: The best step size at point 'x'
    """

    grad_x = grad_function(x)
 
    b1 = 0.5 * np.transpose(grad_x) @ matrix_A @ x
    b2 = 0.5 * np.transpose(x) @ matrix_A @ grad_x 
    b3 = np.transpose(vector_B) @ grad_x
    
    beta = b1 + b2 - b3

    alpha = np.transpose(grad_x) @ matrix_A @ grad_x

    return beta / alpha
```

```python
def polyak_step(x: np.ndarray,
                x_sol: np.ndarray,
                function: callable,
                grad_function: callable,
                alpha: float,
                ) -> float:
    """computes the value of the polyak shore step at a given point. The formula is mentioned in the 

    Args:
        x (np.ndarray): The current point
        x_sol (np.ndarray): The optimal point of the function (assumed to be )
        function (callable): a callable object that computes the value of the function at any given point
        grad_function (callable): a callable object that computes the value of the function's gradient at any given point
        alpha (float): the alpha parameter

    Raises:
        ValueError: if alpha is less than '1'

    Returns:
        float: the value of polyak_shore step at the given point
    """

    if alpha < 1:
        raise ValueError((f"the 'alpha' argument is expected to be at least equal to {1}\n"
                          f"Found: {alpha}"))
    return (function(x) - function(x_sol)) / (alpha * x_difference(grad_function(x), 0) ** 2)
    
```

```python
from typing import Union, List

def gradient_descent(function: callable, 
                     grad_function: callable, 
                     x_0: np.ndarray, 
                     x_sol: np.ndarray,
                     
                     K: int = 10 ** 3,
                     eps: float = 10 ** -5, 
                     mode: str = x_opt_diff,
                     
                     gamma_k: callable = None, 
                     greedy_step_func: callable = None, 
                     polyak_step_func: callable = None,
                     
                     return_history: bool = False
                     ) -> Union[List[np.ndarray], np.ndarray]:
    """This function applies the gradient descent algorithm for any given function

    Args:
        function (callable): a callable object that computes the value of the function at any given point
        grad_function (callable): a callable object that computes the value of the function's gradient at any given point
        x_0 (np.ndarray): the starting point
        x_sol (np.ndarray): The analytical solution of the problem
        K (int, optional): the maximum number of iterations. Defaults to 10**3.
        eps (float, optional): As soon as the criterion goes below this value, the algorithm stops. Defaults to 10**-5.
        mode (str, optional): Which criterion to consider. Defaults to x_opt_diff.
        gamma_k (callable, optional): A callable object that determines the step size at the k-th iteration. Defaults to None.
        greedy_step_func (callable, optional): _description_. Defaults to None.
        polyak_step_func (callable, optional): _description_. Defaults to None.
        return_history (bool, optional): _description_. Defaults to False.

    Raises:
        ValueError: _description_

    Returns:
        Union[List[np.ndarray], np.ndarray]: _description_
    """

    # the first step is to make sure the 'mode' variable is defined above
    if mode not in [x_opt_diff, f_opt_diff, x_diff, f_opt_diff]:
        raise ValueError((f"the variable 'mode' is expected to belong to {[x_opt_diff, f_opt_diff, x_diff, f_opt_diff]}"
                          f"Found: {mode}"))

    # make sure at least one of the step size update methods are passed
    if gamma_k is None and polyak_step_func is None and greedy_step_func is None:
        raise ValueError("The function expects at least one methods to update the step size")

    # find the minimum value of the function   
    f_sol = function(x_sol)

    # a boolean flag to determine whether to stop the iterative process
    criterion_stop = False

    # counter for the number of iterations
    k = 0
    x_current, x_previous = x_0, x_0

    # create a list to save the intermediate points  
    x_history = [x_current]

    while not criterion_stop:
        # save the k-th point
        x_previous = x_current

        # compute the step size        
        if greedy_step_func is not None:
            gamma = greedy_step_func(x_current)

        elif polyak_step_func is not None:
            gamma = polyak_step_func(x_current)
            
        else:
            gamma = gamma_k(k)

        # compute the (k+1)-th point
        x_current = x_current - gamma * grad_function(x_current)  

        # compute the criterion
        if mode == x_opt_diff:
            diff = x_difference(x_sol, x_current)
        elif mode == x_diff:
            diff = x_difference(x_previous, x_current)
        
        elif mode == f_diff:
            diff = f_difference(function(x_previous), function(x_current))
        else:
            diff = f_difference(function(x_current), f_sol)

        # add 'x_current' to the history
        x_history.append(x_current)

        # increment 'k' 
        k += 1

        # stop the iterative process either because the maximum number of iterations is reached 
        # or the difference is less than epsilon
        criterion_stop = diff <= eps or k >= K
    
    assert len(x_history) == k + 1
    
    # the fucntion should return the 'numerical solution': x_current
    return x_history if return_history else x_history[-1]
```

```python
# set the seeds for reproducibility
import random
np.random.seed(69)
random.seed(69)

def problem_func(x: np.ndarray, 
            matrix_A: np.ndarray, 
            vector_B: np.ndarray) -> float:
    """This an implementation of the optimization problem : 0.5 (x^T A x) - b^T x 

    Args:
        x (np.ndarray): a given point
        matrix_A (np.ndarray): the matrix A
        vector_B (np.ndarray): the matrix B

    Returns:
        float: the value of the function given all 3 parameters
    """

    # make sure 'x' and 'B' are converted to column vectors
    x = np.expand_dims(x, axis=-1) if len(x.shape) == 1 else x
    vector_B = np.expand_dims(vector_B, axis=-1) if len(vector_B.shape) == 1 else vector_B

    assert x.shape == (len(x), 1), f"make sure to convert 'x' into a column vector"
    assert vector_B.shape == (len(vector_B), 1), f"make sure to convert 'vector_B' into a column vector"

    result = 0.5 * np.transpose(x) @ matrix_A @ x - np.transpose(vector_B) @ x
    # result is a numpy array, make sure to convert it back to a 'float' value.
    return result.item() 

def problem_grad_func(x: np.ndarray,             
                    matrix_A: np.ndarray = None, 
                    vector_B: np.ndarray = None) -> float:
    """This is an implementation of the gradient of the optimization problem: 0.5 (A + A^T)x - B

    Args:
        x (np.ndarray): a given point
        matrix_A (np.ndarray): the matrix A
        vector_B (np.ndarray): the matrix B

    Returns:
        float: the value of the function gradient given all 3 parameters
    """

    x = np.expand_dims(x, axis=-1) if len(x.shape) == 1 else x
    vector_B = np.expand_dims(vector_B, axis=-1) if len(vector_B.shape) == 1 else vector_B

    assert x.shape == (len(x), 1), f"make sure to convert 'x' into a column vector"
    assert vector_B.shape == (len(vector_B), 1), f"make sure to convert 'vector_B' into a column vector"

    grad = 0.5 * (matrix_A + np.transpose(matrix_A)) @ x - vector_B
    return grad
```

```python
import matplotlib.pyplot as plt

def simulation(dim: int , 
               mu: float, 
               L: float, 
               seed: int = 69,

               K: int = 10 ** 3,
               eps: float = -1, # set to a negative value to avoid stopping the algorithm before 'K' iterations
               mode: str = x_opt_diff, 
               
               gamma_k: callable = None, 
               use_greedy_step: bool = False,
               use_polyak_shore_step: bool = False,
               polyak_alpha: float = None,               
               ) -> Tuple[List[np.ndarray], List[float]]:
    """This function performs the entire simulation / experience
    1. setting up the problem
    2. setting up the step size update method
    3. applying gradient descent
    4. extracting hte different points and computing the needed criterion values 

    Args:
        dim (int): The dimension 'd'
        mu (float): the strong convexity constant
        L (float): The Schiptz constant
        seed (int, optional): The seed to control the randomness of the generaiton process. Defaults to 69.

        
        K (int, optional): the maximum number of iterations. Defaults to 10**3.
        eps (float, optional): As soon as the criterion goes below this value, the algorithm stops. Defaults to 10**-5.
        gamma_k (callable, optional): A callable object that determines the step size at the k-th iteration. Defaults to None.

        use_greedy_step (bool, optional): Whether to use the greedy step. Defaults to False.
        use_polyak_shore_step (bool, optional): Whether to use polyak_shore_step. Defaults to False.
        polyak_alpha (float, optional): the alpha parameter in the polyak_alpha step size update rule. Defaults to None.
            Required if 'use_polyak_shore_step' is set to True

    Returns:
        Tuple[List[np.ndarray], List[float]] : a tuple of the reached points, and their corresponding criterion values
    """


    # the default step size will be of constant value 1 / L
    gamma_k = (lambda _: 1 / L) if gamma_k is None else gamma_k

    A, B, x_sol, x_0 = set_up(dim, mu, L, seed=seed)
    
    # create the function variable
    f = lambda x: problem_func(x, matrix_A=A, vector_B=B)
    grad_f = lambda x: problem_grad_func(x, matrix_A=A, vector_B=B)
    
    # create callable objects for the different steps
    min_step_function = None
    polyak_shore_step_function = None

    # set the step update functions if needed
    if use_greedy_step: 
        min_step_function = lambda x: greedy_step(x, 
                                                matrix_A=A, 
                                                vector_B=B,
                                                grad_function=grad_f)
    elif use_polyak_shore_step:
        if polyak_alpha is None:
            raise ValueError((f"if 'use_polyak_shore_step' is set to {True}, The 'parameter 'polyak_alpha'"
                              f" should be passed as well. \n Found: {polyak_alpha}"))

        polyak_shore_step_function = lambda x: polyak_step(x, 
                                                           x_sol=x_sol, 
                                                           function=f, 
                                                           grad_function=grad_f, 
                                                           alpha=polyak_alpha)


    # get the points from the gradient descent
    x_points = gradient_descent(function=f, 
                                grad_function=grad_f, 
                                x_sol=x_sol, 
                                x_0=x_0,
                                gamma_k=gamma_k,
                                K = K,
                                eps=eps,
                                return_history=True, 
                                mode=mode, 
                                greedy_step_func=min_step_function,
                                polyak_step_func=polyak_shore_step_function
                                )
    
    # estimate the value of the criterions
    criterions = [x_difference(x, x_sol) ** 2 for x in x_points]
    return x_points, criterions   
```

```python
plt.figure(figsize=(10, 10))
def plot_iterations(criterions: Iterable[float],
                    label: int = None, 
                    show: bool = True, 
                    start_index: int = 0):
    if label is None:
        plt.plot(list(range(start_index, len(criterions))), criterions[start_index:])
    else:
        plt.plot(list(range(start_index, len(criterions))), criterions[start_index:], label=str(label))
        
    plt.xlabel('iteration')
    plt.ylabel('criterion (log_{10} scale)')
    
    if show:
        plt.legend()
        plt.show()


for s in np.linspace(start=10, stop=50, num=5):
    x_points, criterions = simulation(dim=50, mu=MU, L=L, K=10 ** 3, seed=int(s), mode=f_opt_diff)
    criterions = [np.log10(c) for c in criterions]
    plot_iterations(criterions=criterions, label=f'x0_{int(s)}', show=False)

plt.legend()
plt.title("||x* - x_k||^2 for different x_0")
plt.show()
```

**NATURE OF CONVERGE**: 

The convergence plots (with different values for $x_0$) suggests that: 

\begin{align*}
log(||x_k - x^{*} ||) & \approx -\alpha k && \text{where $k$ is the number of iterations, $\alpha > 0$ } \\
||x_k - x^{*} || & \approx e^{-\alpha k} && \text{where $k$ is the number of iterations, $\alpha > 0$ } \\
\implies log(||x_k - x^{*} ||) \leq \epsilon ~~ can ~ be ~ estimated ~ as ~   e^{-\alpha k} & \leq \epsilon \\ 
-\alpha k & \leq \log(\epsilon) \\
k &\geq \frac{1}{\alpha} \cdot \log(\epsilon)
\end{align*}

Thus the convergnce is **logarithmic**

### H.1
The theoretical optimal step size $\gamma = \frac{1}{L}$. Nevertheless, practically we can find values $\gamma = \frac{a}{L}$ where $a \in [0.5, 1.5]$ lead to faster convergence.

Additionally, we have $\gamma_k =\frac{\gamma}{\delta + k}$ and $\gamma_k =\frac{\gamma}{\delta + \sqrt{k}}$ and both decreasing with respect to $k$. 

Thus, I suggest the following framework to tune $\gamma, \delta$. Given $a > b \in [0.5, 1.5]$,  I solve the following system:

\begin{cases}
    \frac{\gamma}{\delta + m} = \frac{b}{L}  && \text{where $m$ is the maximum value of the function that updates $k$}\\
    \frac{\gamma}{\delta + 1} = \frac{a}{L}
\end{cases}

The solution is:

\begin{cases}
    \delta =  \frac{bm - a}{a - b}\\
    \gamma =  \frac{a}{L} \cdot (\delta + 1) \\
\end{cases}

Thus we will consider different values of $a$ and $\gamma, \delta$ that satisfy the system:

\begin{cases}
    \frac{\gamma}{\delta + \sqrt{K}} = \frac{1 - a}{L}  \\
    \frac{\gamma}{\delta + 1} = \frac{1 + a}{L}
\end{cases}


```python
plt.figure(figsize=(10, 10))

def gamma_delta(a: float, b: float, m: float):
    assert a > b, "Make sure that 'a' larger than 'b'"
    # first compute delta
    delta = (b * m - a) / (a - b)
    gamma = (a / L) * (delta + 1)

    assert np.isclose(gamma / (delta + m), b / L) 
    assert np.isclose(gamma / (delta + 1), a / L) 

    return gamma, delta

K = 10 ** 3

for diff in np.linspace(start=0.05, stop=0.5, num=11):
    upper, lower = (1 + diff) , (1 - diff)

    gamma, delta = gamma_delta(a=upper, b=lower, m = np.sqrt(K))
    gamma_k = lambda k : (gamma / (delta + np.sqrt(k)))
    
    assert np.isclose(gamma / (delta + 1), upper / L), "first equality does not hold"
    assert np.isclose(gamma / (delta + np.sqrt(K)), lower / L), "second equality does not hold"

    x_points, criterions = simulation(dim=D, mu=MU, L=L, K=K, gamma_k=gamma_k, mode=f_opt_diff, seed=69)
    criterions = [np.log10(c) for c in criterions]
    
    label = f'g_d = {round(gamma, 4)}__{round(delta, 4)}'
    plot_iterations(criterions=criterions, label=label, show=False)

plt.legend()
plt.show()
```

### H.2
To find the best step we need to minimize the function $f(x - \gamma \nabla f)$ with respect to $\gamma$

\begin{align*}
f(x - \gamma \nabla f) &= \frac{1}{2} \cdot (x^T - \gamma \cdot \nabla ^ Tf ) \cdot A \cdot (x - \gamma \cdot \nabla f) - b^T \cdot (x - \gamma \nabla f)\\
&= \frac{1}{2} \cdot [x^T A x - \gamma x^T \cdot A \nabla f - \gamma \cdot \nabla f ^ T A x + \gamma ^ 2 \cdot \nabla ^ T f \cdot A \cdot \nabla f] - b^T x +  b^T \gamma \nabla f\\

\implies \min(f(x - \gamma \nabla f)) &= \min(\frac{1}{2} (\gamma ^ 2 \cdot \nabla ^ T f \cdot A \cdot \nabla f - \gamma(\nabla f ^ T A x + x^T \cdot A \nabla f)) + b^T \gamma \nabla f) \\

\implies \min(f(x - \gamma \nabla f)) &= \min(\frac{1}{2} \gamma ^ 2 \cdot \nabla ^ T f \cdot A \cdot \nabla f - \gamma (\frac{1}{2} \cdot \nabla f ^ T A x + \frac{1}{2} x^T \cdot A \nabla f - b^T\nabla f)) \\

\implies \min(f(x - \gamma \nabla f)) &= \frac{\beta}{\alpha}
\end{align*}

where $\alpha = \nabla ^ T f \cdot A \cdot \nabla f$ and $\beta = (\frac{1}{2} \cdot \nabla f ^ T A x + \frac{1}{2} x^T \cdot A \nabla f - b^T \nabla f)$
