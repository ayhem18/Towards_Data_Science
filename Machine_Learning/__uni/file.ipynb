```python
import datakit.visualize.visualize as vis
import matplotlib.pyplot as plt 
import numpy as np
from itertools import chain

fig, ax = plt.subplots(nrows=1, ncols=1)
d = {"k1": [10, 20, 30], "k2": [30, 40, 50], "k3": [50, 60, 10]}

x = np.arange(3)
bar_width = 0.25

xticks = []

items = list(d.items())
for i, (k, v) in enumerate(items):
	# for i = 1, we need 0, 0.25, 0.5
	x_values = (x + i * len(x)) * bar_width + bar_width * (i != 0) 
	# save the position of the 'x_tick'
	xticks.append(np.median(x_values))

	ax.bar(x_values, v, width=bar_width, align='center', label=k)

ax.set_xticks(xticks, [k for k, v in items])
ax.legend(ncols=len(d))
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)

d = {"k1": [10, 20, 30], "k2": [30, 40, 50], }

x = np.arange(3)

for i, (k, v) in enumerate(d.items()):
	ax.bar(x + i * len(x), v, width=0.25, label=k)

# build the xticks

xticks = [np.median(x + i * len(x)) for i in range(len(d))]

ax.set_xticks(xticks, list(d.keys()))

ax.legend(ncols=len(d))
plt.show()
```
